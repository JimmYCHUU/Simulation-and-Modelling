### **Task-1:** Generate two vectors with 15 random floats each, plot them on the same graph, label the axes, and add a title.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.random.randint(4, 37, size=(4, 4))
B = np.random.randint(1, 29, size=(4, 4))

A_flat = A.ravel()
B_flat = B.ravel()

indices = np.arange(len(A_flat))

plt.plot(indices, A_flat, marker='o', linestyle='-', label="Array 1")
plt.plot(indices, B_flat, marker='s', linestyle='--', label="Array 2")

plt.xlabel('Position')
plt.ylabel('Magnitude')
plt.title('Comparison of Two Arrays')

plt.legend()
plt.grid(True)
plt.show()
```

### **Task-2:** Create a 4x4 matrix with random values, visualize it as a heatmap, and label the rows and columns.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.randint(4, 28, size=(4, 4))

row_names = ['A', 'B', 'C', 'D']
col_names = ['X', 'Y', 'Z', 'W']

sns.heatmap(data, annot=True, cmap='coolwarm', linewidths=0.5)

plt.xticks(np.arange(len(col_names)), col_names)
plt.yticks(np.arange(len(row_names)), row_names)

plt.title('Visual Representation of 4x4 Grid')
plt.show()
```

### **Task-3:** Generate two 4x4 matrices, perform addition, subtraction, and multiplication, visualize the results using bar plots, and discuss the effects.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.randint(1, 10, (4, 4))
Y = np.random.randint(1, 10, (4, 4))

sum_matrix = X + Y
diff_matrix = X - Y
prod_matrix = np.dot(X, Y)

print('Addition Result \n', sum_matrix)
print('Subtraction Result \n', diff_matrix)
print('Dot Product Result \n', prod_matrix)

sum_flat = sum_matrix.flatten()
diff_flat = diff_matrix.flatten()
prod_flat = prod_matrix.flatten()

fig, axes = plt.subplots(1, 3, figsize=(20, 5))

axes[0].bar(range(len(sum_flat)), sum_flat, color='blue')
axes[0].set_title("X + Y")
axes[0].set_ylim([0, 40])

axes[1].bar(range(len(diff_flat)), diff_flat, color='green')
axes[1].set_title("X - Y")
axes[1].set_ylim([-20, 20])

axes[2].bar(range(len(prod_flat)), prod_flat, color='brown')
axes[2].set_title("X * Y")
axes[2].set_ylim([0, 200])

plt.tight_layout()
plt.show()
```

#### **Discussion**
**Matrix Addition:** In the plot of the addition, we can see that each element is simply the sum of the corresponding elements of the two matrices. Since the numbers range from 1 to 9, the result will have values between 2 and 18.

**Matrix Subtraction:** The subtraction plot shows the difference between corresponding elements. The results can be both positive and negative depending on the relative values of A and B.

**Matrix Multiplication:** Matrix multiplication results in a more complex pattern compared to addition and subtraction. The values can vary more dramatically since each element is a sum of products of corresponding elements from rows of A and columns of B. This could lead to much larger values, especially if the original matrices contain large numbers.
